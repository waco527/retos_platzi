## Reto #1 - Suma hasta cincuenta
Crea una variable que se llame total, después pide a tu usuario que ingrese un número y lo sume al valor de total. Haz que esto se repita hasta que el valor de total sea mayor a 50 y muestra el resultado en pantalla.

```python
def intentar (funcion):
    otra_vez = (input('\n¿Quiere intentarlo de nuevo? Si/No ')).lower()
    if otra_vez == 'si' or otra_vez.lower() == 's' :
        funcion()
    elif otra_vez == 'no' or otra_vez.lower() == 'n' :
        print('\nGracias por participar')
    else :
        intentar(funcion)
```

```python
def suma50 ():
    try :
        suma = 0
        total = 0
        while suma < 50 :
            total = float(input('Ingresa un número, esta pregunta se repetira hasta que el valor de la suma de los anteriores números es mayor que 50 '))
            if total < 0 :
                total = 0
                print('debe ingresar un número positivo')
            suma = total + suma
        print ('Acaba de sobrepasar los 50')
    except ValueError :
        print('Ingrese un número valido')
    intentar(suma50)
suma50()
```

## Reto #3 - Sumas consecutivas
Pide al usuario que ingrese dos números y los sume. Después pregunta si quiere añadir otro número: si la respuesta es afirmativa añádelo al total; en caso contrario finaliza el ciclo y muestra el resultado total en pantalla.

```python
def suma_consecutiva():
    try :
        total = 0
        suma = 0
        num1 = float(input('Ingresa dos números para ser sumados. Ingresa el primer nuúmero: '))
        num2 = float(input('Ingresa el segundo número: '))
        total = num1 + num2        
        opcion = (input('¿Quisiera ingresar otro números para ser sumados? S/N')).lower()
        
        while opcion == 's' or opcion == 'si' :
            otro_numero = float(input('Ingrese el número que quiere adicionar: '))
            total = total + otro_numero
            opcion = (input('¿Quisiera ingresar otro números para ser sumados? S/N')).lower()
            
    except ValueError :
        print('Ingrese un número valido')
    print(f'\nEl valor total de la suma es igual a {total}')
    intentar(suma_consecutiva)
suma_consecutiva()
```

## Reto #4 - Lista de invitados
Estás organizando un banquete al que quieres invitar a tus amigos. Crea un programa que pida el nombre de uno de tus amigos, al hacerlo aumenta en uno una variable contadora, después pregunta si quieres invitar a alguien más: si la respuesta es afirmativa entonces pregunta por una persona más; en caso contrario cierra el ciclo y muestra en pantalla cuantos invitados son.

```python
invitados = (input('Programa para organizar un banquete.\nIngrese el nombre de un invitado: ')).capitalize()
i = 1

opcion = (input('¿Quisiera ingresar otro invitado? S/N')).lower()

while opcion == 's' or opcion == 'si' :
    otro_nombre = (input('Ingrese el nombre de otro invitado: \n')).capitalize()
    invitados = invitados + ', \n' + otro_nombre 
    opcion = (input('¿Quisiera ingresar otro invitado? S/N ')).lower()
    i += 1 
print (f'El número de invitados es {i}.\nEsta es la lista de invitados: \n{invitados} ')
```

## Reto #5 - Adivina el número secreto
Crea un programa donde tendrás la variable numero_secreto, la cual toma un valor aleatorio entre 1 y 100. Después pide a tu usuario que indique un número para intentar adivinarlo: en caso de acertar indícale cual era el número secreto y cuantos intentos le tomó; en caso contrario indícale si el número ingresado es mayor o menor al número secreto.

```python
def adivina() :
    import random
    numero_secreto = random.randint(1, 100)
    guess = 0
    try :
        while guess != numero_secreto :
            guess = int(input('Por favor, adivine un número secreto entre 1 y 100.\n'))
            if guess == numero_secreto :
                print(f'Felicitaciones, usted adivinó, el número era {numero_secreto}')
            elif guess < numero_secreto :
                print(f'El número que usted introdujo es menor que el número secreto')
            elif guess > numero_secreto :
                print(f'El número que usted introdujo es mayor que el número secreto')
            else :
                print(f'Introduzca un número valido')
                
    except ValueError :
        print('Debe introducir un número entero')
    intentar(adivina)
    
adivina()
```
